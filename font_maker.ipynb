```python
!apt-get install potrace
```

```python

from tqdm import tqdm
prBar = tqdm(glob.glob(f"{save_bmp}/*.bmp"))
```

```python
len(save_svg)
```

```python
for file in prBar:
    filename = os.path.basename(file)[:-3] + 'svg'
    filepath = os.path.join(save_svg, filename)
    !potrace {file} --svg -o {filepath}
    prBar.set_description(f'filepath : {filepath}')
```

##  SVG TO FONT

```python
import json
```

```python
x = json.load(open('example.json'))
```

```python
len(os.listdir('results/handwritting_fonts_mom_svg'))
```

```python
import os
for idx, i in enumerate(range(0xac00, 0xd7a4)):
    x['glyphs'][hex(i)] = {'src':f'results/handwritting_fonts_mom_svg/{idx}.svg', 'width':128}
```

```python
del x['# vim: set et sw=2 ts=2 sts=2:']
```

```python
del x['glyphs']['0x3f']
del x['glyphs']['0xab']
del x['glyphs']['0x263a']
del x['glyphs']['0x1f304']
del x['glyphs']['0x2723']
```

```python
x['sfnt_names'] = [['Korean', 'Copyright', 'Copyright (c) 2022 by ZolZak'],
                  ['Korean', 'Family', 'Handwrittings mom'],
                  ['Korean', 'SubFamily', 'Regular'],
                  ['Korean', 'UniqueID', 'Handwrittings mom 2022-04-27'],
                  ['Korean', 'Fullname', 'Handwrittings mom Regular'],
                  ['Korean', 'Version', 'Version 001.000'],
                  ['Korean', 'PostScriptName', 'Handwrittings mom-Regular']]
```

```python
x['output'] = ['mom.ttf', 'mom.otf'] # name to convert ttf
```

```python
f = open('handwritting_fonts_mom.json', 'w')
json.dump(x, f)
```
